## Data Cleaning

### Subtask:
Clean the dataset by handling missing values, converting 'Year' to integer, and dropping 'Event_gender' and 'Country_Code' columns.

**Reasoning**:
As per the project's data preparation steps, we need to clean the dataset. This includes checking for and dropping rows with any missing values, converting the 'Year' column from float to integer, and dropping the 'Event_gender' and 'Country_Code' columns as they are either redundant or not directly used in the initial analysis suggested in the PDF. Displaying `df.info()` after cleaning will confirm the changes.

```python
# Check for missing values before cleaning
print("Missing values before cleaning:")
print(df.isnull().sum())

# Drop rows with any missing values
df_cleaned = df.dropna(how='any')

# Convert 'Year' column to integer
df_cleaned['Year'] = df_cleaned['Year'].astype(int)

# Drop 'Event_gender' and 'Country_Code' columns
df_cleaned = df_cleaned.drop(columns=['Event_gender', 'Country_Code'], errors='ignore')

print("\nMissing values after cleaning:")
print(df_cleaned.isnull().sum())

print("\nDataFrame Info after cleaning:")
df_cleaned.info()

print("\nFirst 5 rows of the cleaned DataFrame:")
print(df_cleaned.head())
```

## Analyze Medal Counts by Country

### Subtask:
Calculate the total medals won by each country and visualize the top 10 countries by medal count.

**Reasoning**:
To analyze medal counts by country, I will group the cleaned DataFrame by 'Country' and count the 'Medal' entries for each. Then, I will sort these counts in descending order to identify the top-performing countries and visualize the top 10 using a bar plot, as demonstrated in the PDF.

```python
# 3.1 Total Medal Count by Country
# Total medals won by each country
medals_by_country = df_cleaned.groupby('Country')['Medal'].count().sort_values(ascending=False)

print("Top 10 Countries by Medal Count:")
print(medals_by_country.head(10))

import matplotlib.pyplot as plt

# Plotting the top 10 countries by medals
plt.figure(figsize=(12, 6))
medals_by_country.head(10).plot(kind='bar', color='gold')
plt.title("Top 10 Countries by Medal Count")
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Understand Events and Disciplines

### Subtask:
Identify unique events and analyze event distribution across sports.

**Reasoning**:
To understand the events and disciplines, I will first find the number of unique combinations of 'Sport', 'Discipline', and 'Event' to represent unique events. Then, to analyze the distribution of events across sports, I will group the data by 'Sport' and count the occurrences, similar to the PDF example, and visualize the result using a bar plot.

```python
# Q3. Understand the events themselves.
# Focus on Sport, Discipline and Event. Use groupby and see how many kinds and variations are there.
q3_data_events = df_cleaned[['Sport', 'Discipline', 'Event']].drop_duplicates()
print("Total number of unique events are held so far are :", len(q3_data_events))

# Analyze event distribution across sports
q3_data_sport_distribution = df_cleaned.groupby('Sport')['Event'].nunique().sort_values(ascending=False)
print("\nNumber of unique events per Sport (Top 10):")
print(q3_data_sport_distribution.head(10))

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
q3_data_sport_distribution.head(15).plot(kind='bar', color='lightgreen')
plt.title("Number of Unique Events per Sport (Top 15)")
plt.xlabel("Sport")
plt.ylabel("Number of Unique Events")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Analyze City Hosting Statistics

### Subtask:
Determine which city hosted the maximum unique Olympic years and the most events, and visualize the city event counts.

**Reasoning**:
As per the project requirements, I need to identify cities with the most unique Olympic years and those that hosted the most events. The provided PDF example uses `drop_duplicates` on 'Year' to find unique hosting years and `value_counts()` on 'City' to determine event counts. I will then plot the top cities by event count.

```python
# Q1. Which city hosted the maximum number of unique Olympic years?
# Focus on City and Year. Get unique Year.
q1_data = df_cleaned[['City', 'Year']].drop_duplicates('Year')
print("Cities and their Olympic years:\n", q1_data)

# Count how many times each city appears after dropping duplicate years
city_olympics_count = q1_data['City'].value_counts()
print("\nCity with maximum unique Olympic years:")
print(city_olympics_count.head(1))

# Q2. Which city hosted the most events?
# Focus on City. Find count of unique values (representing events).
q2_data = df_cleaned['City'].value_counts()
print("\nTop 10 Cities by Event Count:")
print(q2_data.head(10))

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
q2_data.head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Cities by Event Count")
plt.xlabel("City")
plt.ylabel("Total Events (Medal Records)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
